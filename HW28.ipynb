```python
#1-3
def classification_func(classification, age): 
    if classification == 'Low' and age < 20: return 'Low_Young'
    elif classification == 'Low' and age > 20: return 'Low_Old'
    elif classification == 'High' and age < 20: return 'High_Young'
    elif classification == 'High' and age >= 20: return 'high_Old'
      
df['Classification_By_Age'] = np.vectorize(classification_func)(df['Classification'], df['Age'])
df
```
